import pandas as pd

organise UN mortality data

```python
pop = pd.read_parquet("../data/processed_data/populated.parquet.gz")
countries = list(set(pop['adm0name']))
mortality = pd.read_csv("../data/raw_data/UNdata_Export_20210603_104133027.csv.gz")
mortality = mortality[mortality['Country or Area'].isin(countries)].reset_index(drop=True)
mortality
```

remove totals

```python
mortality_totals = mortality.query('Month == "Total"')
mortality = mortality.query('Month != "Total"')
mortality_totals
```

```python
mortality = mortality.assign(month=0)
mortality.loc[mortality['Month'] == 'January', 'month'] = 1
mortality.loc[mortality['Month'] == 'February', 'month'] = 2
mortality.loc[mortality['Month'] == 'March', 'month'] = 3
mortality.loc[mortality['Month'] == 'April', 'month'] = 4
mortality.loc[mortality['Month'] == 'May', 'month'] = 5
mortality.loc[mortality['Month'] == 'June', 'month'] = 6
mortality.loc[mortality['Month'] == 'July', 'month'] = 7
mortality.loc[mortality['Month'] == 'August', 'month'] = 8
mortality.loc[mortality['Month'] == 'September', 'month'] = 9
mortality.loc[mortality['Month'] == 'October', 'month'] = 10
mortality.loc[mortality['Month'] == 'November', 'month'] = 11
mortality.loc[mortality['Month'] == 'December', 'month'] = 12
mortality
```

```python
mortality.to_parquet("../data/processed_data/un_mortality.parquet", compression="gzip")
```
