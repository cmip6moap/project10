```python
populated = (shps
    .assign(
        lon=shps['geometry'].x,
        lat=shps['geometry'].y,
        pop=(shps['pop_max']+shps['pop_min'])/2)
    .filter(['name', 'adm0name', 'pop', 'lon', 'lat']))
populated
```

```python
populated.to_parquet('../data/processed_data/populated.parquet.gz', compression='gzip')
```
